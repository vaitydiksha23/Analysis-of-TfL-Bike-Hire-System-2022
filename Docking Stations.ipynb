```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
stations = pd.read_csv('E:\KCL\Project\Cycle Hire 2021\stations 2022.csv')
stations
```

```python
# Change colours for stations based on capacity
def color(capacity):
    if(capacity < 15):
        return('red')
    elif(15 <= capacity < 30):
        return('orange')
    else:
        return('green')

# Create base map
London = [51.506949, -0.122876]
map = folium.Map(location = London,
                 zoom_start = 12, 
                 tiles = "CartoDB dark_matter")
docking_stations = folium.plugins.MarkerCluster().add_to(map)

# Plot markers
for lat, lon, cap, name, id in zip(stations["Latitude"], stations["Longitude"], stations["Docking Station Capacity"], 
                                    stations["Docking Station Name"], stations["Docking Station ID"]):
    folium.CircleMarker(location = [lat, lon], radius = 10, popup = "Capacity: ("+str(cap)+") "+ "ID: ("+str(id)+") "+name,
                    tooltip = str(name), fill_color = color(cap), 
                    fill_opacity= 1, color = "gray").add_to(docking_stations)
map
```
